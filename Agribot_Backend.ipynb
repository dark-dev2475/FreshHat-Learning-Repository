```python
import geemap
```

```python
# Define a point of interest somewhere in India to center our map
# Let's use a point near New Delhi
poi = ee.Geometry.Point([77.216721, 28.644800])

# Get the Landsat 8 Image Collection
collection = (ee.ImageCollection('LANDSAT/LC08/C02/T1_TOA')
              .filterBounds(poi)  # Get images that cover our point
              .filterDate('2023-01-01', '2023-12-31') # From the year 2023
              .filter(ee.Filter.lt('CLOUD_COVER', 20)) # With less than 20% cloud cover
              .sort('CLOUD_COVER') # Sort by the clearest image
              .first()) # Get just the first (clearest) image from the collection

# Check if we got an image
print(collection.getInfo())
```

```python
# Create an interactive map centered on a point that includes Kashmir
map = geemap.Map(center=[34.0, 76.5], zoom=7)

# Define visualization parameters to make it look like a natural color image
vis_params = {
    'bands': ['B4', 'B3', 'B2'], # Red, Green, Blue bands
    'min': 0,
    'max': 0.3
}

# Add the satellite image layer to the map
map.addLayer(collection, vis_params, 'Landsat 8 Image')

# Display the map
map
```

```python
# Line 1:
import ee

# Line 2:
import geemap

# Line 3:
countries = ee.FeatureCollection("FAO/GAUL/2015/level0")

# Line 4:
india_boundary = countries.filter(ee.Filter.eq('ADM0_NAME', 'India'))

# Line 5:
map = geemap.Map(center=[22.5, 82.5], zoom=5)

# Line 6:
map.addLayer(india_boundary, {}, 'India Boundary')

# Line 7:
map
```

```python
# Line 8:
image_collection = ee.ImageCollection('LANDSAT/LC08/C02/T1_TOA')

# Line 9:
start_date = '2020-01-01'
# Line 10:
end_date = '2024-01-01'

# Line 11:
print(f"Searching for images over India from {start_date} to {end_date}...")

# Line 12:
filtered_collection = (image_collection
# Line 13:
                      .filterDate(start_date, end_date)
# Line 14:
                      .filterBounds(india_boundary)
# Line 15:
                      .filter(ee.Filter.lt('CLOUD_COVER', 20)))

# Line 16:
count = filtered_collection.size().getInfo()

# Line 17:
print(f"Success! Found {count} images that match your criteria.")
```
